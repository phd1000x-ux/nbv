A small sample notebook for sanity-checking how `nbv` renders different cell types.

````python
import math

print("Hello from nbv sample!")
````

````python
def fib(n):
    a, b = 1, 1
    out = []
    for _ in range(n):
        out.append(a)
        a, b = b, a + b
    return out

fib(10)
````

````python
for name, val in [("pi", math.pi), ("e", math.e), ("sqrt2", math.sqrt(2))]:
    print(f"{name:5s} ~ {val:.6f}")
````
